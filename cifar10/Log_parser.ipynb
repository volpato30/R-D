```python
%matplotlib inline
import sys
import matplotlib.pyplot as plt
tl = []
vl = []
with open('train_wide_resnet_3_4.log') as read:
    for line in read:
        if line[:4] != 'iter':
            continue
        temp = line.split('|')
        tl.append(float(temp[1][4:]))
        vl.append(float(temp[2][4:]))
```

```python
import matplotlib
import numpy as np
```

```python
matplotlib.style.use('ggplot')
plt.figure(figsize = (20,10))
plt.plot(np.arange(len(tl)),tl,label = 'train loss')
plt.plot(np.arange(len(vl)),vl,'--', label = 'valid loss')
plt.legend(loc='upper right')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('WRN-22-4')
```
